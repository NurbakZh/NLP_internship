Function to divide "text" part of the sentences to words and calculate
average number of characters per words for Raw features 1.

```python
def average_num_of_characters(str):
    lst = str.split(' ')
    av_num = 0
    for word in lst:
        av_num+=len(word)
    av_num /= len(lst)
    return int(av_num)
```

List for Raw Features 1 - **word length**(*average number of characters per words*)
for every sentence in Train-a data.

```python
RF1 = []
for i in range(len(sentences)):
    RF1.append(average_num_of_characters(sentences[i].metadata['text']))
```

```python
#Declaring variables for counting certain upos number
num_adj = 0
num_adp = 0
num_adv = 0
num_aux = 0
num_cconj = 0
num_det = 0
num_intj = 0
num_noun = 0
num_num = 0
num_part = 0
num_pron = 0
num_propn = 0
num_punct = 0
num_sconj = 0
num_sym = 0
num_verb = 0
num_x = 0

#Declaring lists for counting ratios of type/text for every sentence
TTR_adj = []
TTR_adp = []
TTR_adv = []
TTR_aux = []
TTR_cconj = []
TTR_det = []
TTR_intj = []
TTR_noun = []
TTR_num = []
TTR_part = []
TTR_pron = []
TTR_propn = []
TTR_punct = []
TTR_sconj = []
TTR_sym = []
TTR_verb = []
TTR_x = []

for i in range(len(sentences)):
    for word in sentences[i]:
        if(part_of_speech(word) == 'ADJ'):
            num_adj+=1
        elif(part_of_speech(word) == 'ADP'):
            num_adp+=1
        elif(part_of_speech(word) == 'ADV'):
            num_adv+=1
        elif(part_of_speech(word) == 'AUX'):
            num_aux+=1
        elif(part_of_speech(word) == 'CCONJ'):
            num_cconj+=1
        elif(part_of_speech(word) == 'DET'):
            num_det+=1
        elif(part_of_speech(word) == 'INTJ'):
            num_intj+=1
        elif(part_of_speech(word) == 'NOUN'):
            num_noun+=1
        elif(part_of_speech(word) == 'NUM'):
            num_num+=1
        elif(part_of_speech(word) == 'PART'):
            num_part+=1
        elif(part_of_speech(word) == 'PRON'):
            num_pron+=1
        elif(part_of_speech(word) == 'PROPN'):
            num_propn+=1
        elif(part_of_speech(word) == 'PUNCT'):
            num_punct+=1
        elif(part_of_speech(word) == 'SCONJ'):
            num_sconj+=1
        elif(part_of_speech(word) == 'SYM'):
            num_sym+=1
        elif(part_of_speech(word) == 'VERB'):
            num_verb+=1
        elif(part_of_speech(word) == 'X'):
            num_x+=1
    total_words = len(sentences[i].metadata['text'].split(' '))
    TTR_adj.append(num_adj/total_words)
    TTR_adp.append(num_adp/total_words)
    TTR_adv.append(num_adv/total_words)
    TTR_aux.append(num_aux/total_words)
    TTR_cconj.append(num_cconj/total_words)
    TTR_det.append(num_det/total_words)
    TTR_intj.append(num_intj/total_words)
    TTR_noun.append(num_noun/total_words)
    TTR_num.append(num_num/total_words)
    TTR_part.append(num_part/total_words)
    TTR_pron.append(num_pron/total_words)
    TTR_propn.append(num_propn/total_words)
    TTR_punct.append(num_punct/total_words)
    TTR_sconj.append(num_sconj/total_words)
    TTR_sym.append(num_sym/total_words)
    TTR_verb.append(num_verb/total_words)
    TTR_x.append(num_x/total_words)
    #zeroing counters for every sentence
    num_adj = 0
    num_adp = 0
    num_adv = 0
    num_aux = 0
    num_cconj = 0
    num_det = 0
    num_intj = 0
    num_noun = 0
    num_num = 0
    num_part = 0
    num_part = 0
    num_pron = 0
    num_propn = 0
    num_punct = 0
    num_sconj = 0
    num_sym = 0
    num_verb = 0
    num_x = 0
```

Functions below will take mood, tense & person from the words
of sentences, changing lack of such features to "None"

```python
def verbs_mood(word):
    if(word['feats'] is not None):
        if('Mood' in word['feats']):
            return word['feats']['Mood']
    else:
        return "None"

def verbs_tense(word):
    if(word['feats'] is not None):
        if('Tense' in word['feats']):
            return word['feats']['Tense']
    else:
        return "None"

def verbs_num_per(word):
    if(word['feats'] is not None):
        if('Number' in word['feats']):
            return word['feats']['Number']
    else:
        return "None"
```

List for Morpho-Syntactic Features 2 -
**verbal features**(*the distribution of verbs accord-
ing to mood (**verbs mood** - VM), tense (**verbs tense** - VT)
and persons (**verbs num per** - VNP), and lexical density
(**lex density**)* - LD)
for every sentence in Train-a data.

```python
VM = []
VT = []
VNP = []
LD = []

for i in range(len(sentences)):
    ld_counter = 0
    vm_dict = {}
    vt_dict = {}
    vnp_dict = {}
    for word in sentences[i]:
        if(str(verbs_mood(word))!="None"):
            vm_dict[word['form']] = verbs_mood(word)
        if(str(verbs_tense(word))!="None"):
            vt_dict[word['form']] = verbs_tense(word)
        if(str(verbs_num_per(word))!="None"):
            vnp_dict[word['form']] = verbs_num_per(word)
        if((part_of_speech(word) == "VERB") |
        (part_of_speech(word) == "ADJ") |
        (part_of_speech(word) == "ADV") |
        (part_of_speech(word) == "NOUN")):
            ld_counter += 1
    VM.append(vm_dict)
    VT.append(vt_dict)
    VNP.append(vnp_dict)
    LD.append(ld_counter/len(sentences[i].metadata['text'].split(' ')))
```

Train-a Data


Function to determine part of syntactic dependency(**DEPREL**)

```python
def part_of_speech(word):
    return word['deprel']
```

```python
#Determination of types of syntatic dependencies in all sentences
sys_dep = {}
for i in range(len(sentences)):
    for word in sentences[i]:
        if word['deprel'] not in sys_dep:
            sys_dep[word['deprel']] = 0

print(sys_dep)
print(len(sys_dep))
```

List for Syntactic Features 1 -
**probability of syntactic dependency types**
(*subject, direct object, modifier, etc*)
for every sentence in Train-a data.


```python
prob_synt_dep = []
sys_dep_temp = dict(sys_dep)
for i in range(len(sentences)):
    #counter for overall dependency types
    counter = 0
    for word in sentences[i]:
        if word['deprel'] not in sys_dep:
            sys_dep[word['deprel']] = 1
        else:
            sys_dep[word['deprel']] += 1
        counter+=1
    for word in sentences[i]:
        sys_dep[word['deprel']] /= counter
    prob_synt_dep.append(sys_dep)
    sys_dep = dict(sys_dep_temp)

#displaying first element of the list for first sentence
prob_synt_dep[0]
```

Function to determine all sentence roots, that exists

```python
def sentence_roots(sentence):
    roots = {}
    for word in sentence:
        if word['head'] is not None:
            if word['head'] not in roots:
                roots[word['head']] = 1
            else:
                roots[word['head']] += 1
    return roots
```

Function to determine which of the sentence roots are verbal roots.
It creates dictionary of the root type and amount of occurrences
like these:

**{'NOUN': 19, 'PROPN': 1, 'PUNCT': 1, 'VERB': 13}**

Second function below will take this data and count distribution
of verbal roots out of the total sentence roots in form of ratio

```python
def sent_roots(roots,sentence):
    root_types = {}
    for i in roots:
        if sentence[i-1]['upos'] not in root_types:
            root_types[sentence[i-1]['upos']] = roots[i]
        else:
            root_types[sentence[i-1]['upos']] += roots[i]
    return root_types

def sent_roots_ratio(roots):
    sum = 0
    verb_sum = 0
    for i in roots:
        if(i == 'VERB'):
            verb_sum += roots[i]
        sum += roots[i]
    return verb_sum/sum
```

List for Syntactic Features 2 -
**distribution of verbal roots**
(*the distribution of verbal roots out of the total of sentence roots.*)
for every sentence in Train-a data.

```python
DIST_VERB_ROOTS = []
for i in range(len(sentences)):
    DIST_VERB_ROOTS.append(sent_roots_ratio(sent_roots(sentence_roots(sentences[i]), sentences[i])))

DIST_VERB_ROOTS
```

```python
def bfs(graph, root):
    maxdepth = 0
    visited = []
    queue = []
    visited.append(root)
    queue.append((root,1))
    while queue:
        x,depth = queue.pop(0)
        maxdepth = max(maxdepth,depth)
        #print(x)
        if(x in graph):
            for child in graph[x]:
                if child not in visited:
                    visited.append(child)
                    queue.append((child,depth+1))
    return maxdepth-1

#function that creates dictionary with all roots and
#other roots with which it is connected
def sentence_roots_2(sentence):
    roots = {}
    for word in sentence:
        words = []
        if word['head'] is not None:
            if word['head'] not in roots:
                words.append(word['id'])
                roots[word['head']] = words
            else:
                roots[word['head']].append(word['id'])
    return roots

sent = sentence_roots_2(sentences[0])
print("max depth", bfs(sent, 0))
```

List for Syntactic Features 3 -
**parse tree depth features**
(*the depth of the whole parse tree. calculated in terms of the
longest path from the root of the dependency tree
to some leaf*)
for every sentence in Train-a data.

```python
MAX_Depths = []
for sentence in sentences:
    dictionary = sentence_roots_2(sentence)
    MAX_Depths.append(bfs(dictionary,0))
```
